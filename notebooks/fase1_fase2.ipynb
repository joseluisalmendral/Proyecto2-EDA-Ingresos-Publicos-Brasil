## Union Data Frames

## Importamos las librerias que necesitemos

```python
import sys

sys.path.append('../')

from src import support_dataframes as sup_df
import pandas as pd
pd.set_option('display.max_columns', False)
```

## Declaramos nuestros dataframes

```python
df_2013 = pd.read_csv('../datos/datos-2013.csv', sep=';')
df_2014 = pd.read_csv('../datos/datos-2014.csv', sep=';')
df_2015 = pd.read_csv('../datos/datos-2015.csv', sep=';')
df_2016 = pd.read_csv('../datos/datos-2016.csv', sep=';')
df_2017 = pd.read_csv('../datos/datos-2017.csv', sep=';')
df_2018 = pd.read_csv('../datos/datos-2018.csv', sep=';')
df_2019 = pd.read_csv('../datos/datos-2019.csv', sep=';')
df_2020 = pd.read_csv('../datos/datos-2020.csv', sep=';')
df_2021 = pd.read_csv('../datos/datos-2021.csv', sep=';')

lista_dataframes = [df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021]
```

### Limpiamos los nombres de las columnas a minuscula

```python
for df in lista_dataframes:
    df.columns = sup_df.limpiar_lista_palabras(df.columns)
```

## Rellenamos los anios con valores nulos ya que sabemos a que anio pertenecen

```python
anios = [int(i) for i in range(2013, 2022)]
for i, df in enumerate(lista_dataframes):
    df['ano_exercicio'].fillna(anios[i], inplace=True);
    df['ano_exercicio'] = df['ano_exercicio'].astype(int);
```

## Concatenamos los DataFrame

```python
dfs_concatenados = pd.concat(lista_dataframes)

print(dfs_concatenados.shape)
dfs_concatenados.sample(5)
```

## Intentamos rellenar la celda 'nome_orgao_superior' a traves del 'codigo_orgao_superior'

### primero, tratamos los nulos con el valor 999999

```python
dfs_concatenados['codigo_orgao_superior'].fillna(999999, inplace=True)

#? cambiamos la columna a tipo int
dfs_concatenados['codigo_orgao_superior'] = dfs_concatenados['codigo_orgao_superior'].astype(int)

codigos_orgao_superior = dfs_concatenados['codigo_orgao_superior'].unique()
codigos_orgao_superior
```

### Creamos un diccionario con los codigos y nombres de las orgao superiores

```python
diccionario_codigo_nome_superior = {}

for codigo in codigos_orgao_superior:
    condicion = dfs_concatenados['codigo_orgao'] == codigo
    df_codigo_nome = dfs_concatenados[condicion].filter(like='orgao_superior')
    valores = df_codigo_nome['nome_orgao_superior'].unique()
    nome = valores[pd.notna(valores)]

    if len(nome) > 0:
        diccionario_codigo_nome_superior[int(codigo)] = nome[0]
    else:
        diccionario_codigo_nome_superior[int(codigo)] = 'Desconocido'



diccionario_codigo_nome_superior
```

### Ahora cambiamos los nulos

```python
dfs_concatenados['nome_orgao_superior'] = dfs_concatenados['nome_orgao_superior'].fillna(
    dfs_concatenados['codigo_orgao_superior'].map(diccionario_codigo_nome_superior)
)

condicion = dfs_concatenados['codigo_orgao_superior'] == codigos_orgao_superior[3]
dfs_concatenados[condicion]
```

## Vamos a explorar los nulos que tenemos antes de guardar los datos

```python
porcentaje_nulos = sup_df.mostrar_porcentaje_nulos(dfs_concatenados)
porcentaje_nulos
```

## Ahora limpiamos los nombres de orgao con codigo orgao

```python
dfs_concatenados['codigo_orgao'].fillna(999999, inplace=True)

#? cambiamos la columna a tipo int
dfs_concatenados['codigo_orgao'] = dfs_concatenados['codigo_orgao'].astype(int)

codigos_orgao = dfs_concatenados['codigo_orgao'].unique()
codigos_orgao
```

```python
diccionario_codigo_nome = {}

for codigo in codigos_orgao:

    condicion = dfs_concatenados['codigo_orgao'] == codigo
    df_codigo_nome = dfs_concatenados[condicion].filter(regex=r"nome_orgao$")
    valores = df_codigo_nome['nome_orgao'].unique()
    nome = valores[pd.notna(valores)]

    if len(nome) > 0:
        diccionario_codigo_nome[int(codigo)] = nome[0]
    else:
        diccionario_codigo_nome[int(codigo)] = 'Desconocido'

diccionario_codigo_nome
```

```python
dfs_concatenados['nome_orgao'] = dfs_concatenados['nome_orgao'].fillna(
    dfs_concatenados['codigo_orgao'].map(diccionario_codigo_nome)
)

condicion = dfs_concatenados['codigo_orgao'] == codigos_orgao[3]
dfs_concatenados[condicion]
```
