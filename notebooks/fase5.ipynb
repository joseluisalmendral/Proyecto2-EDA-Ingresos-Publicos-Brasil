## 1. Resumen de Hallazgos:

### Identificar las categorías y períodos con mayor discrepancia entre lo previsto y lo realizado.
- Categorías de Capital: Las mayores discrepancias entre ingresos previstos y realizados se observaron en la categoría de Receitas de Capital, especialmente en subcategorías como "Resultado do Banco Central do Brasil" y "Amortizações de Empréstimos", donde la subejecución fue significativa.


- Periodos Críticos: Años como 2013 y 2014 mostraron grandes diferencias en la ejecución de ingresos. Las categorías más volátiles parecen estar influenciadas por factores externos económicos, generando grandes diferencias entre lo presupuestado y lo recaudado.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

dfs_concatenados = pd.read_csv('../datos/tratados/dfs_concatenados.csv')

# Calcular la diferencia entre lo previsto y lo realizado
dfs_concatenados['diferencia'] = dfs_concatenados['valor_previsto_atualizado'] - dfs_concatenados['valor_realizado']

# Agrupar por categoría económica para visualizar las diferencias
ingresos_por_categoria = dfs_concatenados.groupby('categoria_economica').agg({
    'valor_previsto_atualizado': 'sum',
    'valor_realizado': 'sum',
    'diferencia': 'sum'
}).reset_index()

# Gráfico de barras para comparar los ingresos previstos, realizados y las diferencias
plt.figure(figsize=(8, 5))
sns.barplot(x='categoria_economica', y='diferencia', data=ingresos_por_categoria)
plt.title('Discrepancia entre Ingresos Previstos y Realizados por Categoría Económica', fontsize=14)
plt.xlabel('Categoría Económica', fontsize=12)
plt.ylabel('Diferencia (Previsto - Realizado)', fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(True)
plt.show()
```

```python
# Diagrama para mostrar la dispersión de las diferencias entre previsto y realizado
plt.figure(figsize=(12, 8))
sns.boxplot(x='categoria_economica', y='diferencia', data=dfs_concatenados)
plt.title('Dispersión de las Diferencias entre Ingresos Previstos y Realizados por Categoría Económica', fontsize=14)
plt.xlabel('Categoría Económica', fontsize=12)
plt.ylabel('Diferencia (Previsto - Realizado)', fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(True)
plt.show()
```

### Describir tendencias observadas en la ejecución de ingresos.

- Subejecución Generalizada: Se ve una tendencia recurrente de subejecución en categorías como Receitas de Capital y Contribuições Sociais, donde los ingresos realizados no alcanzaron los presupuestos previstos.

- Variabilidad en Receitas Correntes: Aunque algunas subcategorías dentro de Receitas Correntes mostraron una ejecución cercana a lo previsto, otras como Exploração de Recursos Naturais tuvieron discrepancias importantes.
